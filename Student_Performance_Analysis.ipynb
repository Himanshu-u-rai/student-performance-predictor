```python
# Student_Performance_Analysis.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r'H:\Devlopment\StudentData\Student_performance_data _.csv'
data = pd.read_csv(file_path)
```

```python
# Create a binary target variable for classification based on GPA
threshold = 2.0
data['PassStatus'] = data['GPA'].apply(lambda x: 1 if x >= threshold else 0)
```

```python
# Data Visualization
plt.figure(figsize=(14, 6))
```

```python
# Distribution of GPA
plt.subplot(1, 2, 1)
sns.histplot(data['GPA'], bins=20, kde=True)
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
```

```python
# Relationship between StudyTimeWeekly and GPA
plt.subplot(1, 2, 2)
sns.scatterplot(x=data['StudyTimeWeekly'], y=data['GPA'])
plt.title('Study Time Weekly vs GPA')
plt.xlabel('Study Time Weekly')
plt.ylabel('GPA')

plt.tight_layout()
plt.show()
```
